### Llama 3.2-Vision MLLM 

* Mulitmodal Large Language Models (MLLMs) is a collection of pretrained and instruction-tuned image reasoning generative models in 11B and 90B sizes (text + images in / text out). 

* The Llama 3.2-Vision instruction-tuned models are optimized for visual recognition, image reasoning, captioning, and answering general questions about an image.  


### 1. Importing Required Libraries

```python
import ollama
```

### 2. Download the Model

```python
ollama.pull("llama3.2-vision")
```

### Visual Question Answering

```python
response = ollama.chat(
    model="llama3.2-vision",
    messages=[{
       'role': 'user',
       'content': 'what is in this image?',
       'images': ['images\sample.jpg']
    }]
)

print(response['message']['content'])
```

### Visual Question Answering (Streaming)

```python
# create streaming response
stream_response = ollama.chat(
    model="llama3.2-vision",
    messages=[{
       'role': 'user',
       'content': 'can you write a caption for this image?',
       'images': ['images\sample.jpg']
    }],
    stream=True
)


# print chunks in streaming response
for chunk in stream_response:
    print(chunk['message']['content'], end='', flush=True)
```

### OCR (Optical Character Recognition)

```python
# ask model to read the screenshot and convert it to markdown 

stream = ollama.chat(
    model="llama3.2-vision",
    messages=[{
       'role': 'user',
       'content': 'can you read the text in this screenshot and convert it to markdown?',
       'images': ['images\image.png']
    }],
    stream=True
)

for chunk in stream:
    print(chunk['message']['content'], end='', flush=True)
```
